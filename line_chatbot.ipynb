```python
import sys
import requests
from bs4 import BeautifulSoup
import time
import collections
import numpy as np
import os
from PIL import Image
```

```python
rs = requests.session()
```

```python
url_page="https://acc.nctu.edu.tw/?page_id=643"
```

```python
res=rs.get(url_page)
content=res.text
print(content)
soup=BeautifulSoup(content,"html.parser")
content_soup=soup.text
print(content_soup)
```

```python
import re
bull=soup.find_all(class_="has-subtle-pale-blue-background-color has-background")
tmp=0

info=[None,None,None,None,None] #é¡åˆ¥ æ´»å‹•æ™‚é–“ æ´»å‹•åç¨± æ´»å‹•åœ°é» å ±åé€£çµ
f=open('activity_workshop.txt','w')
f2=open('activity_others.txt','w')

for i in bull[0].find_all('td'):   
    tmp=tmp+1
    index=tmp%5-1
    
    if tmp%5!=0:
        title=str(i)
        title2=re.sub(r'<+\w+>','',title)
        title3=re.sub(r'<+\/+\w+>','',title2)
        title4=re.sub(r'<+\w+\/+>','',title3)
        info[index]=title4
    else:
        title=i.find_all('a')
        if len(title)>0:
            title2=title[0].get('href')
            info[index]=title2
            
            print(info[0]+"\n"+info[2]+"\n"+"æ—¥æœŸ : "+info[1]+"\n"+"åœ°é» : "+info[3]+"\n"+info[4]+"\n")
            print("\n")
            
            if info[0]=="å·¥ä½œåŠ":
                f.write(info[2]+"\n"+"æ—¥æœŸ : "+info[1]+"\n"+"åœ°é» : "+info[3]+"\n"+info[4]+"\n"+"\n")
            else:
                f2.write(info[2]+"\n"+"æ—¥æœŸ : "+info[1]+"\n"+"åœ°é» : "+info[3]+"\n"+info[4]+"\n"+"\n")
        else:
            title=str(i)
            title2=re.sub(r'<+\w+>','',title)
            title3=re.sub(r'<+\/+\w+>','',title2)
            title4=re.sub(r'<+\w+\/+>','',title3)
            info[index]=title4
            
f.close()
f2.close()
```
